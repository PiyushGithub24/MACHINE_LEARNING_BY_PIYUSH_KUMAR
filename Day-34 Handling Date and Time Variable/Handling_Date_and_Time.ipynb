#Lets work with dates :--

```python
date["date"]=pd.to_datetime(date["date"])
```

```python
date.info()
```

##Extract year :--

```python
date["date_year"]=date["date"].dt.year
date.sample(5)
```

##Extract month :--

```python
date["date_month_no"]=date["date"].dt.month
date.sample(5)
```

##Extract month name:--

```python
date["date_month_name"]=date["date"].dt.month_name()
date.sample(5)
```

##extract day :--

```python
date["date_day"]=date["date"].dt.day
date.sample(5)
```

##EXtract day of week ,ie,sunday monday ...

```python
date["date_day_of_week"]=date["date"].dt.dayofweek
date.sample(5)
```

##Extract day of week name,ie,1,2,3,4,5,6 or 7.

```python
date["date_day_of_week_name"]=date["date"].dt.day_name()
date.sample(5)
```

```python
date.drop(columns=["city_id","product_id","orders"]).sample(5)
```

#Is weekend :--

```python
date["date_is_weekend"]=np.where(date["date_day_of_week_name"].isin(["Sunday","Saturday"]),1,0)
```

```python
date.drop(columns=["city_id","product_id","orders"]).sample(5)
```

##Extract week of the year :--

```python
date["date_week"]=date["date"].dt.isocalendar().week
```

```python
date.drop(columns=["city_id","product_id","orders"]).sample(5)
```

##Extract quarter :--

```python
date["date_quarter"]=date["date"].dt.quarter
```

```python
date.drop(columns=["city_id","product_id","orders"]).sample(5)
```

#Extract semester :--

```python
date["date_semester"]=np.where(date["date_quarter"].isin([1,2]),1,2)
```

```python
date.drop(columns=["city_id","product_id","orders"]).sample(5)
```

##Exract elapsed time b/w dates :--

```python
import datetime
today=datetime.datetime.today()
today
```

```python
today-date["date"]
```

```python
(today-date["date"]).dt.days
```

#month passed :--

```python
np.round( (today-date["date"]) /np.timedelta64(1,"M"), 0)
```

#Lets work with time :--

```python
time
```

```python
#converting to datetime datatype
time["date"]=pd.to_datetime(time["date"])
```

```python
time.info()
```

##extract hours ,minutes and seconds :--

```python
time["hours"]=time["date"].dt.hour
time["minutes"]=time["date"].dt.minute
time["seconds"]=time["date"].dt.second
```

```python
time.sample(5)
```

```python
time["time"]=time["date"].dt.time
```

```python
time.sample(5)
```

#time difference :--

```python
today-time["date"]
```

```python
#in hours
np.round((today-time["date"])/np.timedelta64(1,"h"),2)
```

```python
#in minutes
np.round((today-time["date"])/np.timedelta64(1,"m"),0)
```

```python
#in second
np.round((today-time["date"])/np.timedelta64(1,"s"),0)
```
