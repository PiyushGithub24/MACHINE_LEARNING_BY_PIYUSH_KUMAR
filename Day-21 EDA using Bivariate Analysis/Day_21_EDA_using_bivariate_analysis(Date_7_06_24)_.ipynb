#1.Scatterplot(Numerical- Numerical) :--

#4.Distplot(Numerical - Categorical) :-

```python
sns.distplot(titanic['Age'],color='green')
```

```python
#instead of the above i can use below code
sns.histplot(
    titanic["Age"], kde=True,
    stat="density", kde_kws=dict(cut=3)
)
```

```python
sns.kdeplot(titanic[titanic['Survived']==0]['Age'],color='green',fill=True)
sns.kdeplot(titanic[titanic['Survived']==1]['Age'] ,color='red')
plt.legend(titanic['Survived'].unique())
```

#5.Heatmap(Categorical - Categorical) :--

```python
titanic.head(3)
```

```python
titanic.info()
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
mean_survived_per_pclass = titanic.groupby('Pclass')['Survived'].mean()*100
print(mean_survived_per_pclass)
```

```python
titanic.groupby('Sex')['Survived'].mean()*100
```

```python
titanic.groupby('Embarked')['Survived'].mean()*100
```

```python
mean_survived_per_pclass.plot(kind='bar')
plt.title('Bar Plot :--',color='red')
plt.xlabel('Pclass :--',color='green')
plt.ylabel('Average Survived :--',color='green')

#or we can use this

# sns.barplot(mean_survived_per_pclass,color='green')
# plt.title('Bar Plot :--',color='red')
```
