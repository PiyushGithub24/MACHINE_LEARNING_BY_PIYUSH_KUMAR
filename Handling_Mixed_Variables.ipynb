```python
titanic["number_numerical"]=pd.to_numeric(titanic["number"],errors="coerce",downcast="integer")
titanic["number_categoricall"]=np.where(titanic["number_numerical"].isnull(),titanic["number"],np.nan)
```

```python
titanic["Cabin_numerical"]=titanic["Cabin"].str.extract('(\d+)')   #captures numerical part
titanic["Cabin_categoricall"]=titanic["Cabin"].str[0] #captures the first part
```

```python
#extract first part of ticket as number
titanic["ticket_num"]=titanic["Ticket"].apply(lambda x:x.split()[-1])
titanic["ticket_num"]=pd.to_numeric(titanic["ticket_num"],errors="coerce", downcast="integer")

#extract second part of ticket as number
titanic["ticket_cat"]=titanic["Ticket"].apply(lambda x:x.split()[0])
titanic["ticket_cat"]=np.where(titanic["ticket_cat"].str.isdigit(),np.nan,
                               titanic["ticket_cat"])

titanic.head(10)
```
