#Logistic Regression Using Gradient Descent :-

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=20)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(penalty=None,solver='sag')
lr.fit(X,y)
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
m1 = -(lr.coef_[0][0]/lr.coef_[0][1])
b1 = -(lr.intercept_/lr.coef_[0][1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m1*x_input + b1
```

```python
np.insert(X,0,1,axis=1)
```

```python
np.ones(X.shape[1])
```

```python
def gd(X,y):

    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.5

    for i in range(5000):
        y_hat = sigmoid(np.dot(X,weights))
        weights = weights + lr*(np.dot((y-y_hat),X)/X.shape[0])

    return weights[1:],weights[0]
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
coef_,intercept_ = gd(X,y)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input1 + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3, label="SKlearn LR")
plt.plot(x_input1,y_input1,color='black',linewidth=3, label="Gradient Descent LR")
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
plt.legend(loc='upper right')
plt.show()
```
