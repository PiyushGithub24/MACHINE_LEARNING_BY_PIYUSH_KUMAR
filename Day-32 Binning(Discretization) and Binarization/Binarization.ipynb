#Binarization :--

##Importing the dependencies :--

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

from sklearn.preprocessing import Binarizer
from sklearn.compose import ColumnTransformer
```

```python
df=pd.read_csv("train.csv",usecols=["Age","Fare","SibSp","Parch","Survived","Survived"])       #Wheather this dataset is not as approprite to apply Binarization .
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df["Family"]=df["SibSp"]+df["Parch"]
df.head()
```

```python
df.drop(columns=["SibSp","Parch"] ,inplace=True)
df.head()
```

```python
X=df.drop(columns=["Survived"])
Y=df["Survived"]
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
```

#Applying Decision Tree without Binarization :--

```python
clf=DecisionTreeClassifier()
clf.fit(X_train,Y_train)
y_pred=clf.predict(X_test)
print("Accuracy score of DT : ",accuracy_score(Y_test,y_pred))
```

```python
clf=DecisionTreeClassifier()
scores=cross_val_score(clf,X,Y,cv=10,scoring="accuracy")
print("Cross validation score  : ",scores  )
print("Cross validation score mean : ",scores.mean())
```

#Applying Decision Tree with using Binarization :--

```python
trf=ColumnTransformer([
    ("Binarizer",Binarizer(),["Family"])
    ],remainder="passthrough")
```

```python
X_train_trf=trf.fit_transform(X_train)
X_test_trf=trf.transform(X_test)
```

```python
clf=DecisionTreeClassifier()
clf.fit(X_train_trf,Y_train)
y_pred=clf.predict(X_test_trf)
print("Accuracy score of DT : ",accuracy_score(Y_test,y_pred))
```

```python
clf=DecisionTreeClassifier()
trf=ColumnTransformer([
    ("Binarizer",Binarizer(),["Family"])
    ],remainder="passthrough")
X_trf=trf.fit_transform(X)
scores=cross_val_score(clf,X_trf,Y,cv=10,scoring="accuracy")
print("Cross validation score  : ",scores  )
print("Cross validation score mean : ",scores.mean())
```
