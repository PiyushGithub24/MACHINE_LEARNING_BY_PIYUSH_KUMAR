```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv('train.csv')
```

```python
train.head()
```

#Categorical Data:--

##(a).countplot:--

```python
sns.countplot(data=train,x='Survived',hue='Survived')
train['Survived'].value_counts()
```

```python
train['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(data=train,x='Pclass',hue='Pclass')
```

```python
sns.countplot(data=train,x='Sex',hue='Pclass')
```

```python
sns.countplot(data=train,x='Embarked',hue='Embarked')
```

##(b).pie chart :--

```python
sex_count=train['Sex'].value_counts()
sex_count.plot(kind='pie',autopct='%.2f',legend=True)
plt.xlabel('Sex')
plt.ylabel('Sex',color='red')
plt.title('Distribution of sex :-',color='red')
#plt.legend(labels=sex_count.index)     another way
```

```python
plt.figure(figsize=(10,5))
train['Survived'].value_counts().plot(kind='pie',autopct='%.2f',legend=True)
plt.ylabel('Survived',color='red')
plt.title('Distribution of Survived :-',color='red')

plt.figure(figsize=(10,5))
train['Embarked'].value_counts().plot(kind='pie',autopct='%.2f',legend=True)
plt.ylabel('Embarked',color='red')
plt.title('Distribution of Embarked :-',color='red')
```
