#Feature Construction :-

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('train.csv')[['Age','Pclass','SibSp','Parch','Survived']]
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

##cross_val_score

```python
np.mean(cross_val_score(LogisticRegression(),X,y,cv=20))*100
```

#Applying feature construction :-

```python
X["family_size"]=df["SibSp"]+df["Parch"]+1
```

```python
X.head()
```

```python
def myfun(num):
  if num==1:
    #"alone"
    return 0
  elif num>=2 and num<=4:
    #"small"
    return 1
  else:
    #"large"
    return 2
```

```python
X["family_type"]=X["family_size"].apply(myfun)
```

```python
X.head()
```

```python
X.drop(["SibSp","Parch","family_size"],axis=1,inplace=True)
```

```python
X.head()
```

```python
np.mean(cross_val_score(LogisticRegression(),X,y,cv=20))*100 #accuracy improves
```
