```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/placement.csv")
```

```python
print(df.head())
print(df.shape)
```

#Steps:--

1>Preprocess +EDA +Feature selection

2>Extracting the input and output column

3>Scale the values

4>train-test-split

5>train the model

6>Evaluate the model/model selection

7>Deployement of the model

#1>Preprocess +EDA +Feature selection :-

```python
df.info()
```

```python
df=df.iloc[:,1:]
df.head()
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['cgpa'],df['iq'],c=df['placement'])
plt.xlabel("cgpa",color='red',fontsize=20)
plt.ylabel("iq",color='red',fontsize=20)
```

#2>Extracting the input and output column :-

```python
x=df.iloc[:,0:2]
y=df.iloc[:,-1]
print(x)
print(x.shape)
print(y)
print(y.shape)
```

#4>train-test-split :-

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
print(type(x_train))
print(type(x_test))
print(type(y_train))
print(type(y_test))
```

```python
x_train
```

#3>Scale the values:-

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
print(x_train)
print('\n',x_test )
```

#5>train the model:-
- since our data is linearly seperable so logistic regression is best to apply on this dataset.

```python
from sklearn.linear_model import LogisticRegression
LRM=LogisticRegression()
LRM.fit(x_train,y_train)
```

```python
y_pred=LRM.predict(x_test)
print("Predicted output for test data:-",y_pred)
print("   Actual output for test data:-",np.array(y_test))
```

#6>Evaluate the model/model selection:-

```python
from sklearn.metrics import accuracy_score
print("Accuracy Score:-",accuracy_score(y_test,y_pred))
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x_train,y_train.values,clf=LRM,legend=2)
plt.xlabel("cgpa",color='red',fontsize=20)
plt.ylabel("iq",color='red',fontsize=20)
plt.show()
```
