#Gradient Descent Step By Step :--

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,Y=make_regression(n_samples=4,n_features=1,n_informative=1,noise=80 ,n_targets=1,random_state=13)
```

```python
X.shape ,Y.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,Y)
```

##Lets first apply OLS :-

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X,Y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red')
```

##Lets apply Gradient Descent assuming slope is constant m=78.35
##and lets assume the starting value of intercept b=0

```python
#Iteration 1 :-
y_pred=(78.35*X + 0).reshape(4)
```

```python
plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='green',label="b=0")
plt.legend()
plt.show()
```

```python
m=78.35
b=0

loss_slope=-2*np.sum(Y-m*X.ravel()-b)
loss_slope
```

```python
#Lets take learning rate lr=0.1
lr=0.1
step_size=lr*loss_slope
step_size
```

```python
#calculating the new intercept
b=b-step_size
b
```

```python
#Iteration 2 :-
y_pred1=(78.35*X + b).reshape(4)

plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred,color='limegreen',label="b=0")
plt.plot(X,y_pred1,color='green',label="b=20.927763408209216")
plt.legend()
plt.show()
```

```python
#iteration:3
loss_slope=-2*np.sum(Y-m*X.ravel()-b)
loss_slope
```

```python
step_size=lr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred2=(78.35*X + b).reshape(4)

plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='darkgreen',label="b=25.11331608985106")
plt.plot(X,y_pred1,color='green',label="b=20.927763408209216")
plt.plot(X,y_pred,color='limegreen',label="b=0")
plt.legend()
plt.show()
```

```python
#iteration 4:-
loss_slope=-2*np.sum(Y-m*X.ravel()-b)
loss_slope
```

```python
step_size=lr*loss_slope
step_size
```

```python
b=b-step_size
b
```

```python
y_pred3=(78.35*X + b).reshape(4)

plt.figure(figsize=(12,10))
plt.scatter(X,Y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='blue',label="b=25.95042662617943")
plt.plot(X,y_pred2,color='darkgreen',label="b=25.11331608985106")
plt.plot(X,y_pred1,color='green',label="b=20.927763408209216")
plt.plot(X,y_pred,color='limegreen',label="b=0")
plt.legend()
plt.show()
```

##if Learning rate lr=0.01:-

```python
b=-100
m=78.35
lr=0.01
epoch=10

for i in range(epoch):
  loss=-2*np.sum(Y-m*X.ravel()-b)
  step_size=lr*loss
  b=b-step_size
  y_pred=m*X+b

  plt.plot(X,y_pred)

plt.plot(X,reg.predict(X),color='red')
plt.scatter(X,Y)
plt.show()
```
