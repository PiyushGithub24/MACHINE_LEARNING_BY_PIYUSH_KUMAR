```python
import numpy as np
import pandas as pd
```

```python
customer=pd.read_csv("customer.csv")
customer.sample(5)
```

##in the given dataframe :
- ordinal variable : review and education (inputs) (apply OrdinalEncoder)
- nominal variable : gender and purchased (inputs) (apply OneHotEncoder)
- discrete numerical variable : age       (outputs) (apply LabelEncoder)

```python
customer_new=customer.iloc[:,2:]
customer_new.sample(5)
```

#train_test_split :-

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(customer_new.iloc[:,:2],customer_new.iloc[:,-1],test_size=0.2,random_state=0)
```

```python
x_train.shape , x_test.shape , y_train.shape , y_test.shape
```

```python
x_train
```

```python
y_train
```

#OrdinalEncoder :-

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
oe.fit(x_train)
oe.categories_

# x_train[['review','education']]=oe.fit_transform(x_train[['review','education']])
# x_test[['review','education']]=oe.transform(x_test[['review','education']])
# x_train=x_train.astype('int')
# x_test=x_test.astype('int')
```

```python
x_train=oe.transform(x_train)
x_test=oe.transform(x_test)
x_train
```

#LabelEncoder :-

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_train=le.fit_transform(y_train)
y_test=le.transform(y_test)
y_train
```
