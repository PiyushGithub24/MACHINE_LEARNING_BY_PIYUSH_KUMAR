# Naive Baye's Machine Learning Algorithms Implementation:

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
X,y=load_iris(return_X_y=True)
```

```python
X
```

```python
y
```

### Train test split:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
```

```python
X_train.shape ,X_test.shape,y_train.shape,y_test.shape
```

### Importing and creating object of gausian naive baye's :

```python
from  sklearn.naive_bayes import GaussianNB
gaussian=GaussianNB()
```

```python
gaussian.fit(X_train,y_train)
```

```python
y_pred=gaussian.predict(X_test)
```

```python
y_pred
```

### Checking the accuracy of the model :

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
print()
print(accuracy_score(y_pred,y_test))
print()
print(classification_report(y_pred,y_test))
```

### encoding on categorical columns:
- sex    -->label encoding
- time   -->label encoding
- smoker -->label encoding
- day    -->one hot encoding

```python
from sklearn.preprocessing import LabelEncoder
le1=LabelEncoder()
le2=LabelEncoder()
le3=LabelEncoder()
```

```python
X_train['sex'] = le1.fit_transform(X_train['sex'])     
X_train['time'] = le2.fit_transform(X_train['time'])
y_train = le3.fit_transform(y_train)
```

```python
X_train
```

```python
X_test['sex'] = le1.fit_transform(X_test['sex'])     
X_test['time'] = le2.fit_transform(X_test['time'])
y_test = le3.fit_transform(y_test)
```

### day -->one hot encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
ct=ColumnTransformer([('OHE',OneHotEncoder(drop='first'),['day'])],
                     remainder='passthrough')
```

```python
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
X_train=ct.fit_transform(X_train)
```

```python
X_train
```

```python
X_test=ct.transform(X_test)
X_test
```

## trainig the Gaussian Naive Bayes algorithm :

```python
from sklearn.naive_bayes import GaussianNB
gaussian=GaussianNB()
```

```python
gaussian.fit(X_train,y_train)
```

```python
y_pred=gaussian.predict(X_test)
y_pred
```

## Checking the accuracy of the model :

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_pred,y_test))
print()
print(accuracy_score(y_pred,y_test))
print()
print(classification_report(y_pred,y_test))
```
