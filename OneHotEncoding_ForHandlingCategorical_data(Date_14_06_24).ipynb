#OneHotEncoding for handling categorical data :-

```python
#importing some libraries
import numpy as np
import pandas as pd
```

```python
cars=pd.read_csv('cars.csv')
cars.head()
```

##here in the given cars dataframe:
- nominal categorical variables : brand ,fuel and owner
##So that's why we will use here OneHotEncoding

```python
#let's see the different categories in brand :
cars['fuel'].value_counts()
```

```python
print(f"Number of different categories in 'brand'=",cars['brand'].nunique())
print(f"Number of different categories in 'fuel'=",cars['fuel'].nunique())
print(f"Number of different categories in 'owner'=",cars['owner'].nunique())
```

#1. OneHotEncoding using Pandas

```python
dum=pd.get_dummies(cars,columns=['fuel','owner'])
dum
```

```python
bool_var=dum.columns[dum.dtypes=='bool']
dum[bool_var]=dum[bool_var].astype(int)
dum.head()
```

#2. K-1 OneHotEncoding :

```python
dum=pd.get_dummies(cars,columns=['fuel','owner'],drop_first=True)
dum
```

```python
bool_var=dum.columns[dum.dtypes=='bool']
dum[bool_var]=dum[bool_var].astype(int)
dum.head()
```

#3. OneHotEncoding using Sklearn :-

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(cars.iloc[:,0:4],cars.iloc[:,-1],test_size=0.2,random_state=2)
```

```python
x_train.shape , y_train.shape , x_test.shape , y_test.shape
```

```python
x_train
```

```python
#importing OneHotEncoder :
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first',sparse_output=False,dtype=np.int32)
```

```python
x_train_new=ohe.fit_transform(x_train[['fuel','owner']])
x_test_new=ohe.transform(x_test[['fuel','owner']])
x_train_new
```

```python
x_train_new.shape
```

```python
x_train[['brand','km_driven']].values
```

```python
np.hstack((x_train[['brand','km_driven']].values,x_train_new))
```

#4. OneHotEncoding with Top Categories :-

```python
counts = cars['brand'].value_counts()
counts
```

```python
threshold = 100
cars['brand'].nunique()
```

```python
repl = counts[counts <= threshold].index
repl
```

```python
dum=pd.get_dummies(cars['brand'].replace(repl, 'uncommon'))
dum.sample(5)
```

```python
bool_var=dum.columns[dum.dtypes=='bool']
dum[bool_var]=dum[bool_var].astype(int)
dum.sample(5)
```
