#Importing some libraries:-

```python
import pandas as pd
from pandas_datareader import data
import numpy as np
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import r2_score
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import GridSearchCV
```

```python
calif=fetch_california_housing()
df=pd.DataFrame(calif.data,columns=calif.feature_names)
```

```python
df['MedHouseVal']=calif.target
df
```

```python
#split data into i/p and o/p
x=df.drop('MedHouseVal',axis=1)
y=df['MedHouseVal']
```

```python
x.shape,y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape,x_test.shape
```

```python
dtr=DecisionTreeRegressor(criterion='squared_error',max_depth=5)
dtr.fit( x_train,y_train)
```

```python
y_pred=dtr.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

#Hyperparameter tunning:-

```python
param_grid={
    'criterion':['squared_error','absolute_error'],
    'max_depth':[1,4,8,10,None],
    'max_features':[0.25,0.5,1.0],
    'min_samples_split':[0.25,0.5,1.0]

}
```

```python
reg=GridSearchCV(DecisionTreeRegressor(),param_grid=param_grid)
```

```python
reg.fit(x_train,y_train)
```

```python
reg.best_score_
```

```python
reg.best_params_
```
