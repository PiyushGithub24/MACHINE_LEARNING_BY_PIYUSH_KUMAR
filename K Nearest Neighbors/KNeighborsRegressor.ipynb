# K Nearest Neighbors Classifier :

```python
#creating dataset:
from sklearn.datasets import make_regression
X, y = make_regression(n_samples=2000, n_features=2, noise=30, random_state=2)
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
#importing and creating object of KNeighborsRegressor
from sklearn.neighbors import KNeighborsRegressor
KNR=KNeighborsRegressor(n_neighbors=6,algorithm='auto')
```

```python
KNR.fit(X_train,y_train)
```

```python
y_pred=KNR.predict(X_test)
y_pred
```

### Check the accuracy of the model :

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
r2_scores=r2_score(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)

print(f"r2_scores :{r2_scores}")
print(f"MAE :{mae}")
print(f"MSE :{mse}")
```
