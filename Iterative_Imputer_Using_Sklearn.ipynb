#Iterative Imputer Using Sklearn :-

```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LinearRegression
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
```

```python
df = np.round(pd.read_csv('50_Startups.csv')[['R&D Spend','Administration','Marketing Spend','Profit']]/10000)
np.random.seed(9)
df = df.sample(5)
df
```

```python
df = df.iloc[:,0:-1]
df
```

```python
df.iloc[1,0] = np.NaN
df.iloc[3,1] = np.NaN
df.iloc[-1,-1] = np.NaN
```

```python
df.head()
```

```python
lr=LinearRegression()
imputer = IterativeImputer(estimator=lr,max_iter=2,random_state=9)
```

```python
imputer
```

```python
imputer.fit_transform(df)
```

```python
np.array(df)
```

```python
def iter_impute(df,n_iter):
  lr=LinearRegression()
  imputer1 = IterativeImputer(estimator=lr,max_iter=n_iter-1,random_state=9)
  imputer2 = IterativeImputer(estimator=lr,max_iter=n_iter,random_state=9)
  return imputer2.fit_transform(df)-imputer1.fit_transform(df)
```

```python
iter_impute(df,6)
```
